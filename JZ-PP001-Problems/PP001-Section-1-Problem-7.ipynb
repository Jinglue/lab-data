# 折叠列表

> 题目来自 CodeWars [Fold an array](https://www.codewars.com/kata/fold-an-array)

完成函数 `fold(lst, n)`，将整数列表 `lst` 折叠 `n` 次，返回折叠后的列表。折叠方式：

```
折叠一次:
[1,2,3,4,5] -> [6,6,3]
```

```
折叠两次:
[1,2,3,4,5] -> [9,6]
```

请从 `# YOUR CODE HERE` 下方另起一行填写你的答案。

**注意**，不要删除 `# YOUR CODE HERE` 注释。

````python
def fold(lst, n):
    ### BEGIN SOLUTION
    lst_len = len(lst)
    
    if lst_len == 1: return lst
    if        n == 0: return lst
    
    folded_lst = []
    for index in range((lst_len + 1) // 2):
        if index == lst_len - index - 1:
            folded_lst.append(lst[index])
        else:
            folded_lst.append(lst[index] + lst[-index-1])
    return fold(folded_lst, n-1)
    ### END SOLUTION
````
