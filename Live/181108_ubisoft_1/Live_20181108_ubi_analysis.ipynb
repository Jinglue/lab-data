## Sales and Ratings

https://www.kaggle.com/kendallgillies/video-game-sales-and-ratings

The fields include:

- Name - The game's name
- Platform - Platform of the games release
- Year_of_Release - Year of the game's release
- Genre - Genre of the game
- Publisher - Publisher of the game
- NA_Sales - Sales in North America (in millions)
- EU_Sales - Sales in Europe (in millions)
- JP_Sales - Sales in Japan (in millions)
- Other_Sales - Sales in the rest of the world (in millions)
- Global_Sales - Total worldwide sales (in millions)
- Critic_score - Aggregate score compiled by Metacritic staff
- Critic_count - The number of critics used in coming up with the critic score
- User_score - Score by Metacritic's subscribers
- User_count - Number of users who gave the user score
- Rating - The ESRB ratings

# Web Scrawler

https://www.metacritic.com/game

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.metacritic.com/company/ubisoft'

params = {
    'filter-options':'games',
    'num_items':100,
    'sort_options':'date',
    'page':18
}
```

```python
headers = {'User-Agent':'user-agent:Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.26 Safari/537.36 Core/1.63.6776.400 QQBrowser/10.3.2577.400'}
```

```python
response = requests.get(url=url, params=params, headers=headers)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'lxml')
```

```python
game_list = soup.find(class_='company_credits').tbody.find_all('tr')
```

```python
game_list[0].find(class_='title').a.text
```

```python
game_list[0].find(class_='year').text.strip()
```

```python
game_list[0].find(class_='score').text.strip()
```

```python
from tqdm import tqdm_notebook
```

```python
games = []
release_dates = []
scores = []

for page in tqdm_notebook(range(19)):
    params['page'] = page
    response = requests.get(url=url, params=params, headers=headers)
    soup = BeautifulSoup(response.text, 'lxml')
    game_list = soup.find(class_='company_credits').tbody.find_all('tr')
    
    for g in game_list:
        games.append(g.find(class_='title').a.text)
        release_dates.append(g.find(class_='year').text.strip())
        scores.append(g.find(class_='score').text.strip())
```

```python
ubi_df = pd.DataFrame({
    'Title':games,
    'Date':release_dates,
    'Score':scores
})
```

```python
# ubi_df.to_csv('UBI_game_rates.csv')
ubi_df = pd.read_csv('UBI_game_rates.csv')
```

```python
valid_ubi_df = ubi_df[ubi_df['Score'].str.contains('\d')][ubi_df['Date'] != 'Canceled'][ubi_df['Date'] != 'TBA']
```

```python
valid_ubi_df['Score'] = pd.to_numeric(valid_ubi_df['Score'])
valid_ubi_df['Date'] = pd.to_datetime(valid_ubi_df['Date'])
```

```python
import re
```

```python
valid_ubi_df['Title'] = valid_ubi_df['Title'].apply(lambda x: re.sub(' \(\w+\)', '', x))
```

```python
ubi_rating_df = valid_ubi_df.groupby(['Title']).max()
```

```python
ubi_games = ubi_rating_df.index
ubi_dates = ubi_rating_df['Date']
ubi_rates = ubi_rating_df['Score']
```

```python
from pyecharts import Scatter
```

```python
scatter = Scatter("Ubisoft Games")
```

```python
scatter.add('Score', ubi_dates, ubi_rates, is_visualmap=True, label_text_color='#FFF',
            xaxis_type='time', is_datazoom_show=True)
```
