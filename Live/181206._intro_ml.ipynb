# 1 特征与标签 features & labels

```python
X = load_iris()['data'][:, 2:]
y = load_iris()['target']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title('Iris dataset')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(X[:,0], X[:,1], c=y, cmap='jet', alpha=0.8, linewidths=0)
```

# 2. 决策树分类

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
dt_clf = DecisionTreeClassifier(max_depth=2)
```

```python
dt_clf.fit(X, y)
```

```python
import pydotplus
from IPython.display import Image  
```

```python
dot_data = export_graphviz(dt_clf, out_file=None, 
                            feature_names=load_iris().feature_names[2:],  
                            class_names=load_iris().target_names,  
                            filled=True, rounded=True,  
                            special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)  
```

```python
Image(graph.create_png())
```

## 2.1 模型评估

```python
y_pred = dt_clf.predict(X)
```

```python
accuracy_score(y, y_pred)
```

```python
from sklearn.datasets import load_digits
```

```python
images = load_digits()['images']
```

```python
m,n = 3,10

fig, axes = plt.subplots(m, n)
for i in range(m*n):
    ax = axes[i//n, i%n]
    ax.imshow(images[i], cmap='gray_r')
    ax.axis('off')
```

```python
X = load_digits()['data']
y = load_digits()['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp_clf = MLPClassifier(hidden_layer_sizes=[32], activation='relu')
```

```python
mlp_clf.fit(X_train, y_train)
```

```python
test_images = X_test.reshape((-1, 8, 8))
```

```python
y_pred = mlp_clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
m,n = 3,10

fig, axes = plt.subplots(m, n)
for i in range(m*n):
    ax = axes[i//n, i%n]
    ax.set_title(y_pred[i])
    ax.imshow(test_images[i], cmap='gray_r')
    ax.axis('off')
```
